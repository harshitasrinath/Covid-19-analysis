# Covid-19 Data Analysis Notebook

#### To import modules

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
print('Modules successfully imported!')
```

#### To import the dataset

```python
corona_df = pd.read_csv('covid19_Confirmed_dataset.csv')
corona_df.head()
```

#### To check the shape of the dataframe

```python
corona_df.shape
```

From the table, we can observe that the latitude and longitude columns won't be of any use. So let us drop them.

#### To delete useless columns

```python
corona_df.drop(["Lat","Long"],axis=1, inplace=True)
```

```python
corona_df.head(10)
```

#### To aggregrate the rows by the country

```python
corona_df_agg = corona_df.groupby("Country/Region").sum()
```

```python
corona_df_agg.head()
```

#### Visualizing data related to a country

```python
corona_df_agg.loc["China"].plot()
corona_df_agg.loc["Italy"].plot()
corona_df_agg.loc["Spain"].plot()
plt.legend()
```

```python
corona_df_agg.loc["China"][:3].plot()
```

#### To calculate and plot the first derivative of the curve

```python
corona_df_agg.loc["China"].diff().plot()
```

#### To find maximum infection rate for China, Italy and Spain

```python
corona_df_agg.loc["China"].diff().max()
```

```python
corona_df_agg.loc["Italy"].diff().max()
```

```python
corona_df_agg.loc["Spain"].diff().max()
```

#### To find maximum infection rate for all the coutries

```python
countries = list(corona_df_agg.index)
max_infection_rates= []
for c in countries :
    max_infection_rates.append(corona_df_agg.loc[c].diff().max())
corona_df_agg["max_infection_rate"] = max_infection_rates
```

```python
corona_df_agg.head()
```

```python
corona_data = pd.DataFrame(corona_df_agg["max_infection_rate"])
```

```python
corona_data.head()
```

```python
corona_data.shape
```

#### To import the world happiness report dataset

```python
happiness_report_df = pd.read_csv("worldwide_happiness_report.csv")
```

```python
happiness_report_df.head()
```

#### To drop useless columns

```python
useless_cols = ["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
```

```python
happiness_report_df.drop(useless_cols, axis=1, inplace=True)
happiness_report_df.head()
```

#### To change the indices of the dataframe

```python
happiness_report_df.set_index("Country or region", inplace=True)
```

```python
happiness_report_df.head()
```

```python
happiness_report_df.shape
```

```python
data = corona_data.join(happiness_report_df, how="inner")
data.head()
```

#### Correlation matrix

```python
data.corr()
```

## Visualization of the results

```python
x = data["GDP per capita"]
y = data["max_infection_rate"]
sns.scatterplot(x, np.log(y))
```
